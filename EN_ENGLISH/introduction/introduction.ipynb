# Introduction to artificial intelligence.

Artificial intelligence is a vast field of computer science and mathematics.

There are 2 main families in artificial intelligence: symbolic AI and statistical AI.

## Symbolic AI

Symbolic AI is an approach that focuses on the manipulation of symbols and rules. It uses explicit models of knowledge and logical reasoning techniques to solve problems.

### What is a symbol?

A symbol represents knowledge in various forms such as words, phrases, concepts, or objects that can be manipulated by algorithms.

Symbolic AI is typically represented in the form of graphs, decision trees, or sets of logical rules.

## Statistical AI

Statistical AI is an approach that enables machines to learn how to solve problems through databases. There are different ways to implement machine learning.

### Supervised learning

Supervised learning AI uses annotated data, meaning that the AI knows in advance what the data should represent.

For example:

- It will know in advance if the input image for its training is a dog or a cat.

### Unsupervised learning

Unsupervised learning AI uses simple unannotated databases. The AI does not know in advance the expected response with the data.

For example:

- We provide a series of photographs of a circle to an algorithm. It will try to understand what a circle is by generalizing the characteristics of a circle until it can recognize them.
  
- In the example of unsupervised AI speech recognition, a system learns to convert speech into text by analyzing audio recordings without specific instructions. It detects patterns of sounds and tones, forming an understanding of speech structures that allows it to transcribe what it hears, adapting to various accents and dialects.

### Self-supervised learning

Self-supervised learning uses unlabeled (unannotated) data. Artificial intelligence models must predict the structures/patterns in the data in order to create labels for this data.

For example:

- Translation tools often use this type of model. Indeed, they will learn patterns/representations of sentences in different languages.

### Reinforcement learning

Reinforcement learning is a specific branch of machine learning where there is an agent that learns to make decisions by interacting with an environment.

What is an agent?

- It is an entity that makes decisions (for example, a robot) in an environment (for example, a path with obstacles).

The agent will perform actions in this given environment and receive rewards or penalties based on the actions it takes.

For example:

- A bot that learns to play Tetris.
