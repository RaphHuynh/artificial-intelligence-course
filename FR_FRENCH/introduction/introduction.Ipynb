# Introduction à l'intelligence artificielle

L'intelligence artificielle est un domaine vaste de l'informatique et des mathématiques.

Il existe 2 grandes familles dans l'intelligence artificielle. Les IA symboliques et les IA statistiques.

## Les IA symboliques

L'intelligence artificielle symbolique est une approche qui se concentre sur la manipulation de symboles et de règles. Elle utilise des modèles explicites de connaissances et des techniques de raisonnement logique pour résoudre des problèmes.

### Qu'est ce qu'un symbole ?

Un symbole représente les connaissances sous divers formes comme des mots, phrases, des concepts ou des objets qui peuvent être manipulés par des algorithmes.

Les IA symboliques sont généralement représentées sous forme de graphes, d'arbres de décision ou encore sous forme d'ensembles de règles logiques.

## Les IA statistiques

L'intelligence artificielle statistique est une approche qui permet d'apprendre aux machines à résoudre des problèmes via des bases de données. Il existe différente manière de mettre en place l'apprentissage machine.

### Apprentissage supervisé

Les IA par apprentissage supervisées utilisent des données dites annotées, c'est à dire que l'ia sait à l'avance ce que doit représenter la donnée à la fin.

Par exemple :

- Il saura en avance si l'image fournie en entrée pour son entrainement est un chien ou un chat.

### Apprentissage non supervisé

Les IA par apprentissage non supervisé utilisent de simple base de données non annotées. L'IA ne sait donc pas à l'avance la réponse attendu avec la donnée.

Par exemple :

- On donne une série de photographies d'un cercle à un algorithme. Il va tenter de comprendre ce qu'est un cercle en généralisant les caracteristiques d'un cercle jusqu'à ce qu'il puisse les repérer.
  
- Dans l'exemple de la reconnaissance vocale par IA non supervisée, un système apprend à convertir la parole en texte en analysant des enregistrements audio sans directives spécifiques. Il détecte des motifs de sons et de tonalités, formant une compréhension des structures de la parole qui lui permet de transcrire ce qu'il entend, s'adaptant à divers accents et dialectes.

### Apprentissage auto-supervisé

L'apprentissage auto-supervisée utilise des données non étiquetées (annotées). Les modèles d'intelligence artificielle doivent prédire eux même les structures / motifs dans les données afin de créer des étiquettes à ces données. 

Par exemple :

- Les outils de traduction utilisent souvent ce type de modèle. En effet, ils vont apprendre des patterns / représentations de phrases dans différentes langues.

### Apprentissage par renforcement

L'apprentissage par renforcement est une branche spécifique de l'apprentissage automatique où il y a un agent qui apprend à prendre des décisions en intéragissant avec un environnement.

Qu'est qu'un agent ?

- C'est une entité qui va prendre des décisions (par exemple un robot) dans un environnement (par exemple un parcours avec des obstacles) 

L'agent va donc effectuer des actions dans cette environnement donné et recevoir des récompenses ou des malus en fonction des actions qu'il effectue.

Par exemple :

- Un bot qui va apprendre à jouer à Tetris.
